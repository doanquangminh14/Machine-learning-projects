```python
df['gender'] = df['gender'].replace(['Female','Male'],[0,1])
```

```python
df = pd.get_dummies(df,columns = ['country'], drop_first= False)
```

```python
df = df.drop('customer_id', axis = 1)
```

```python
df.info()
```

```python
select_columns = df.select_dtypes(include= ['bool'])
from sklearn.preprocessing import LabelEncoder
for columns in select_columns:
    df[columns] = LabelEncoder().fit_transform(df[columns])
```

```python
df
```

```python
X = df.drop('churn', axis = 1, inplace = False).values
y = df['churn'].values
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 0)
```

```python
from imblearn.over_sampling import SMOTE
X_train,y_train = SMOTE().fit_resample(X_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
acc = []
for i in range (2,20):
    model_tree = DecisionTreeClassifier(max_depth=i)
    model_tree.fit(X_train,y_train)
    y_pred = model_tree.predict(X_test)
    acc.append(accuracy_score(y_test,y_pred))
    
```

```python
plt.plot(range(2,20),acc)
plt.xticks(range(2,20))
plt.axvline(x = 6, linestyle = '--', color = 'red')
plt.grid()
```

```python
model_tree = DecisionTreeClassifier(max_depth=6)
model_tree.fit(X_train,y_train)
y_pred = model_tree.predict(X_test)
ac = accuracy_score(y_test,y_pred)
ac
```

```python
accc = []
for i in range (2,20):
    from sklearn.ensemble import RandomForestClassifier
    model_rf = RandomForestClassifier(n_estimators= i)
    model_rf.fit(X_train,y_train)
    y_pred = model_rf.predict(X_test)
    accc.append(accuracy_score(y_test,y_pred))
```

```python
plt.plot(range(2,20),accc)
plt.xticks(range(2,20))
plt.axvline(x = 18, linestyle = '--', color = 'red')
plt.grid()
```

```python
model_rf = RandomForestClassifier(n_estimators= 18)
model_rf.fit(X_train,y_train)
y_pred = model_rf.predict(X_test)
a = accuracy_score(y_test,y_pred)
a
```

```python
from sklearn.model_selection import GridSearchCV
space = {
           'max_depth':[5,10,15],
           'n_estimators':[10,20,30],
           'criterion':['gini','entropy']
}
grid_rf = GridSearchCV(RandomForestClassifier(),param_grid=space, scoring='accuracy')
grid_rf.fit(X_train,y_train)
```

```python
grid_rf.best_params_
```

```python
y_pred = grid_rf.predict(X_test)
a = accuracy_score(y_test,y_pred)
a
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
y_probabilities = grid_rf.predict_proba(X_test)[:,1]
false_positive_rate, true_positive_rate, thresold = roc_curve(y_test,y_probabilities)
plt.plot(false_positive_rate,true_positive_rate, color = 'red')
plt.text(0.4,0.2,'AUC : {:.2f}'.format(roc_auc_score(y_test,y_probabilities)))
```
